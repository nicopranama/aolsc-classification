```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Data produksi tas bulanan dari Januari 2018 hingga Desember 2023
produksi_data = [
    1863, 1614, 2570, 1685, 2101, 1811, 2457, 2171, 2134, 2502, 2358, 2399,
    2048, 2523, 2086, 2391, 2150, 2340, 3129, 2277, 2964, 2997, 2747, 2862,
    3405, 2677, 2749, 2755, 2963, 3161, 3623, 2768, 3141, 3439, 3601, 3531,
    3477, 3376, 4027, 3175, 3274, 3334, 3964, 3649, 3502, 3688, 3657, 4422,
    4197, 4441, 4736, 4521, 4485, 4644, 5036, 4876, 4789, 4544, 4975, 5211,
    4880, 4933, 5079, 5339, 5232, 5520, 5714, 5260, 6110, 5334, 5988, 6235,
    6365, 6266, 6345, 6118, 6497, 6278, 6638, 6590, 6271, 7246, 6584, 6594,
    7092, 7326, 7409, 7976, 7959, 8012, 8195, 8008, 8313, 7791, 8368, 8933,
    8756, 8613, 8705, 9098, 8769, 9544, 9050, 9186, 10012, 9685, 9966, 10048,
    10244, 10740, 10318, 10393, 10986, 10635, 10731, 11749, 11849, 12123,
    12274, 11666, 11960, 12629, 12915, 13051, 13387, 13309, 13732, 13162,
    13644, 13808, 14101, 13992, 15191, 15018, 14917, 15046, 15556, 15893,
    16388, 16782, 16716, 17033, 16896, 17689
]

# Membuat array bulan dari M1 hingga M144
bulan = np.arange(1, 145).reshape(-1, 1)

# Menggunakan Polynomial Regression
poly = PolynomialFeatures(degree=5)
bulan_poly = poly.fit_transform(bulan)

# Fit model polynomial regression
model = LinearRegression()
model.fit(bulan_poly, produksi_data)

# Dapatkan koefisien polinomial
coefficients = model.coef_
intercept = model.intercept_

# Menampilkan koefisien model
print("Koefisien Polinomial:")
for i, coef in enumerate(coefficients):
    print(f"a_{i}: {coef}")
print(f"a_0 (intercept): {intercept}")

# Prediksi data
prediksi_production = model.predict(bulan_poly)

# Evaluasi model
mse = mean_squared_error(produksi_data, prediksi_production)
r2 = r2_score(produksi_data, prediksi_production)

# Visualisasi hasil
plt.figure(figsize=(10, 6))
plt.plot(bulan, produksi_data, marker='o', linestyle='-', label='Produksi Bulanan Asli')
plt.plot(bulan, prediksi_production, color='red', linestyle='--', label='Prediksi Produksi')
plt.title('Grafik Produksi Bulanan Tas dengan metode Polynomial Regression')
plt.xlabel('Bulan')
plt.ylabel('Produksi')
plt.legend()
plt.grid(True)
plt.show()

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
